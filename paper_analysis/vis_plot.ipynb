```python
import matplotlib.pyplot as plt
import scienceplots

# 스타일 설정
plt.style.use('science')
plt.rcParams['font.weight'] = 'bold'
# 데이터 설정
noise_levels = ['0.00', '0.01', '0.03', '0.05', '0.15']
psnr_values_1 = [19.64, 16.60, 14.43, 13.59, 11.89]
psnr_values_2 = [21.88, 16.43, 13.74, 12.78, 10.19]
psnr_values_ours = [19.09] * len(noise_levels)  # 모든 노이즈 레벨에서 동일한 값

# 그래프 생성 (가로:세로 비율 1:2)
fig, ax = plt.subplots(figsize=(4, 6), dpi=300)  # 가로 3, 세로 6으로 비율을 1:2로 설정

# 노이즈 레벨 vs PSNR 플롯 (첫 번째 라인: PixelSplat)
ax.plot(range(len(noise_levels)), psnr_values_1, marker='o', linestyle='-', color='navy', label='PixelSplat')

# 노이즈 레벨 vs PSNR 플롯 (두 번째 라인: MVSplat)
ax.plot(range(len(noise_levels)), psnr_values_2, marker='s', linestyle='--', color='darkred', label='MVSplat')

# 노이즈 레벨 vs PSNR 플롯 (Ours, 세 번째 라인)
ax.plot(range(len(noise_levels)), psnr_values_ours, marker='^', linestyle=':', color='forestgreen', label='Ours')

# 축 라벨 설정, Noise Level (\sigma)에서 실제 기호 사용, fontsize 옵션 적용
# ax.set_xlabel(r'Noise Level ($\sigma$)', fontsize=20, weight='bold')  # x축 글씨 크기 및 bold 설정
# ax.set_ylabel('PSNR (dB)', fontsize=20, weight='bold')    

# 제안된 x축 눈금만 표시하고 간격을 동일하게 설정
ax.set_xticks(range(len(noise_levels)))
ax.set_xticklabels(noise_levels, fontsize=16, weight='bold')  #
ax.tick_params(axis='y', labelsize=14)                  # y축 눈금 레이블 글씨 크기 설정

# 축 범위와 눈금 조정
ax.set_xlim(-0.1, len(noise_levels)-0.9)  # 동일한 간격을 위해 x축 범위를 설정
ax.set_ylim(10, 22)   # PSNR 범위

# 그리드 제거
ax.grid(False)

# 범례 추가 (글씨 크기 설정)
ax.legend(fontsize=20)

# 레이아웃 및 그래프 저장 (PDF 형식으로)
plt.tight_layout()
plt.savefig('psnr_vs_noise_1_to_2.png', format='png')  # PDF로 저장
plt.show()
```
