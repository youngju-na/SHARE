```python
import matplotlib.pyplot as plt
import scienceplots

#  
plt.style.use('science')
plt.rcParams['font.weight'] = 'bold'
#  
noise_levels = ['0.00', '0.01', '0.03', '0.05', '0.15']
psnr_values_1 = [19.64, 16.60, 14.43, 13.59, 11.89]
psnr_values_2 = [21.88, 16.43, 13.74, 12.78, 10.19]
psnr_values_ours = [19.09] * len(noise_levels)  #     

#   (:  1:2)
fig, ax = plt.subplots(figsize=(4, 6), dpi=300)  #  3,  6  1:2 

#   vs PSNR  (  : PixelSplat)
ax.plot(range(len(noise_levels)), psnr_values_1, marker='o', linestyle='-', color='navy', label='PixelSplat')

#   vs PSNR  (  : MVSplat)
ax.plot(range(len(noise_levels)), psnr_values_2, marker='s', linestyle='--', color='darkred', label='MVSplat')

#   vs PSNR  (Ours,   )
ax.plot(range(len(noise_levels)), psnr_values_ours, marker='^', linestyle=':', color='forestgreen', label='Ours')

#   , Noise Level (\sigma)   , fontsize  
# ax.set_xlabel(r'Noise Level ($\sigma$)', fontsize=20, weight='bold')  # x    bold 
# ax.set_ylabel('PSNR (dB)', fontsize=20, weight='bold')    

#  x     
ax.set_xticks(range(len(noise_levels)))
ax.set_xticklabels(noise_levels, fontsize=16, weight='bold')  #
ax.tick_params(axis='y', labelsize=14)                  # y     

#    
ax.set_xlim(-0.1, len(noise_levels)-0.9)  #    x  
ax.set_ylim(10, 22)   # PSNR 

#  
ax.grid(False)

#   (  )
ax.legend(fontsize=20)

#     (PDF )
plt.tight_layout()
plt.savefig('psnr_vs_noise_1_to_2.png', format='png')  # PDF 
plt.show()
```
